```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('numbers')
plt.ylabel('squares')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

```python
import numpy as np
t=np.arange(1,10,0.2)
```

```python
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,'g^',label='-')
plt.plot(t,t**2.5,'rs',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5.0)# For specifying the width
plt.show()
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[1,3,5,6]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[1],color='r',linewidth=3)
plt.setp(lines[0],color='g',linewidth=5)
plt.grid()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.show()
```

```python
plt.figure(2)
plt.plot([4,5,6])
plt.show()
```

```python
x=(1,2,3,4)
y=(1,2,9,16)
plt.scatter(x,y)# change the plot as per lineplot,bar,scatter
plt.show()
```

```python
y=np.arange(0,10,2)
plt.hist(y)
plt.show()
```
