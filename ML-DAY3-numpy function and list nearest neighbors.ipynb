```python
train=np.random.randint(10,100,(100,4))#Training data set
```

```python
train[:,0]=np.random.randint(0,99,100)# making the col more realistic age
```

```python
train[:,1]=np.random.randint(10000,100000,100)#making the col more realistic salary
```

```python
train[:,2]=np.random.randint(1,20,100)#making the col more realistic Location
```

```python
train[:,3]=np.random.randint(0,2,100)#making the col more realistic EMI or not
```

```python
train.shape
```

```python
trainDS=train[0:80,:]
```

```python
trainDS.shape
```

```python
testDS=train[80:100,:]
```

```python
testDS.shape
```

```python
trainDS=trainDS[:,0:3]
```

```python
trainDS.shape
```

```python
holdout=testDS[0,0:3]#Who comes newly to the system, not psrt of training DS
```

```python
holdout.shape
```

```python
d=(holdout-trainDS)*(holdout-trainDS)
```

```python
d1=d.sum(axis=1)
```

```python
np.sqrt(d1)
```

```python
d1.argmin()
```
