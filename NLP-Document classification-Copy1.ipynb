```python
import PyPDF2
import glob
import os
finaldataset=[]
for i in os.listdir("D:\\Users\\abhisv\\Downloads\\databases"):
    for j in glob.glob("D:\\Users\\abhisv\\Downloads\\databases\\"+i+"\\*.pdf"):
        pdfFileObject=open(j,'rb')
        pdfReader=PyPDF2.PdfFileReader(pdfFileObject)
        text=''
        for k in range(0,pdfReader.numPages):
            pageObj=pdfReader.getPage(k)
            text=text+pageObj.extractText()
            print(i,j,k)
        finaldataset.append([i,j,text])
            
```

```python
import pandas as pd
finaldataset=pd.DataFrame(finaldataset)
```
